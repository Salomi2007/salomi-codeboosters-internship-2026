```python
import pandas as pd
df=pd.read_csv("/content/sample_data/student_performance.csv")
df.head()
```

```python
print("no of student:",df.shape[0])
print("no of columns:",df.shape[1])
print(f"Unique Departments       : {df['department'].nunique()}")
```

```python
print("\nDEPARTMENT WISE ANALYSIS")
dept_count = df['department'].value_counts()
print(dept_count)
```

```python
print("\nSCORE ANALYSIS")
print(f"Highest Math Score        : {df['math_score'].max()}")
print(f"Lowest Math Score         : {df['math_score'].min()}")
print(f"Average Math Score        : {df['math_score'].mean():.2f}")
print(f"Average Programming Score : {df['programming_score'].mean():.2f}")
print(f"Average Attendance        : {df['attendance_percentage'].mean():.2f}")
```

```python
df['total_score']=df['math_score']+df['science_score']+df['programming_score']+df['english_score']
print("\nGENDER WISE ANALYSIS")
gender_avg = df.groupby('gender')['total_score'].mean()
print(gender_avg)
```

```python
print("\nTOP 5 PERFORMERS")
top_students = df.sort_values(by='total_score', ascending=False).head(5)
for index, row in top_students.iterrows():
    print(f"\nStudent Name : {row['name']}")
    print(f"Department   : {row['department']}")
    print(f"Total Score  : {row['total_score']}")
```

```python
print("\nATTENDANCE ANALYSIS")
low_attendance = df[df['attendance_percentage'] < 75]
if len(low_attendance) > 0:
    for index, row in low_attendance.iterrows():
        print(f"\nName        : {row['name']}")
        print(f"Department  : {row['department']}")
        print(f"Attendance  : {row['attendance_percentage']}%")
else:
    print("All students have attendance above 75%")
```

```python
electronics_students = len(df[df['department'] == 'Electronics'])

print("Students from Electronics Department :", electronics_students)
```

```python
female_attendance = df[df['gender'] == 'Female']['attendance_percentage'].mean()

print("Average Attendance of Female Students :", female_attendance)
```

```python
lowest_programming = df.loc[df['programming_score'].idxmin()]

print("Student with Lowest Programming Score :", lowest_programming['name'])
```

```python
above_80_math = len(df[df['math_score'] > 80])

print("Students Scored Above 80 in Math :", above_80_math)
```

```python
high_attendance_avg = df[df['attendance_percentage'] > 90]['total_score'].mean()

print("Average Total Score (Attendance > 90%) :", high_attendance_avg)
```
